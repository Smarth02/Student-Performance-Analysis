#Student Performance Analysis
This project analyzes how study habits, lifestyle factors, attendance, mental health, and other student characteristics relate to academic performance

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
url="https://raw.githubusercontent.com/Smarth02/Student-Performance-Analysis/main/student_habits_performance.csv"
df=pd.read_csv(url) #Read dataset
print(df.head()) #First 5 rows of the datset
print(df.info()) #Information about the dataset
print(df.describe()) #Statistical summary of the dataset
```

```python
print('Number of Null values')
print(df.isna().sum())
```

```python
df=df.dropna() #Remove all rows which have null value(s)
df.isna().sum() #Number of null values
```

```python
print(df['part_time_job'].unique()) #Unique values in part_time_job
print(df['diet_quality'].unique()) #Unique values in diet_quality
print(df['extracurricular_participation'].unique()) #Unique values in extracurricular_participation
print(df['internet_quality'].unique()) #Unique values in internet_quality
```

```python
#Replace strings with integers
df['part_time_job']=df['part_time_job'].replace({'Yes':1, 'No':0})
df['diet_quality']=df['diet_quality'].replace({'Fair':2, 'Good':1, 'Poor':0})
df['extracurricular_participation']=df['extracurricular_participation'].replace({'Yes':1, 'No':0})
df['internet_quality']=df['internet_quality'].replace({'Good':2, 'Average':1, 'Poor':0})
```

```python
print('First five rows of dataset')
df.head()
```

```python
#Select numerical columns
numerical_df=df.select_dtypes(include=np.number)
#Correlation matrix
corr_matrix=numerical_df.corr()
#Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Number of students by gender
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gender', palette='viridis')
plt.title('Number of students by gender')
plt.xlabel('Gender')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on whether they do a part time job or not
sns.countplot(data=df, x='part_time_job', palette='viridis')
plt.title('Number of students based on whether they do a part time job or not')
plt.xlabel('Yes or No')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on diet quality
sns.countplot(data=df, x='diet_quality', palette='viridis')
plt.title('Number of students based on diet quality')
plt.xlabel('Diet Quality')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on exercise frequency
sns.countplot(data=df, x='exercise_frequency', palette='viridis')
plt.title('Number of students based on exercise frequency')
plt.xlabel('Exercise Frequency')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on parental education level
sns.countplot(data=df, x='parental_education_level', palette='viridis')
plt.title('Number of students based on parental education level')
plt.xlabel('Parental Education Level')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on internet quality
sns.countplot(data=df, x='internet_quality', palette='viridis')
plt.title('Number of students based on internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Number of students')
plt.show()
```

```python
#Number of students based on extracurricular participation
sns.countplot(data=df, x='extracurricular_participation', palette='viridis')
plt.title('Number of students based on extracurricular participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Number of students')
plt.show()
```

```python
#Variation of marks with Study hours per day
plt.figure(figsize=(8, 5))
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=df, color='red')
plt.title('Variation of Exam score with study hours per day')
plt.xlabel('Study hours per day')
plt.ylabel('Exam Score')
plt.show()
```

```python
colors=sns.color_palette('pastel')
```

```python
#Average exam score by gender
gender_scores=df.groupby('gender')['exam_score'].mean()
print(gender_scores)
plt.bar(gender_scores.index, gender_scores.values, color=colors)
plt.title('Average exam score by gender')
plt.ylabel('Average Exam Score')
plt.xlabel('Gender')
plt.show()
```

```python
#Average exam score by internet quality
internet_scores=df.groupby('internet_quality')['exam_score'].mean()
print(internet_scores)
plt.bar(internet_scores.index, internet_scores.values, color=colors)
plt.title('Average exam score by internet quality')
plt.xlabel('Internet Quality')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
#Average exam score by extracurricular participation
extracurricular_scores=df.groupby('extracurricular_participation')['exam_score'].mean()
print(extracurricular_scores)
plt.bar(extracurricular_scores.index, extracurricular_scores.values, color=colors)
plt.title('Average Exam score by extracurricular participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
#Variation of exam score with important parameters
cols=['social_media_hours', 'netflix_hours', 'attendance_percentage', 'sleep_hours', 'exercise_frequency', 'mental_health_rating']
for col in cols:
  sns.scatterplot(x=col, y='exam_score', data=df, color='green')
  plt.title(col+' vs Exam Score')
  plt.xlabel(col)
  plt.ylabel('Exam Score')
  plt.show()
```

```python
#Box plots to detect outliers
box_columns=['gender', 'part_time_job', 'diet_quality', 'parental_education_level', 'internet_quality']
for column in box_columns:
  plt.figure(figsize=(8, 4))
  sns.boxplot(data=df, x=column, y='exam_score', palette='Set2')
  plt.title('Exam score vs '+column)
  plt.xlabel(column)
  plt.ylabel('Exam Score')
  plt.xticks(rotation=20)
  plt.show()
```
